# **Task - Image to pencil sketch with Python**
**Level - Beginner Level**

**Importing required libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**Loading the original image**

```python
image = cv2.imread("Bird1.png")
cv2_imshow(image)
```

**Converting RGB Image into grayscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

**Inverting the gray scale image**

```python
invert_img = cv2.bitwise_not(gray_image)
cv2_imshow(invert_img)
```

**Smoothen the inverted image using gaussian blur technique**

```python
img_smoothen = cv2.GaussianBlur(invert_img, (21, 21), sigmaX = 0, sigmaY = 0)
cv2_imshow(img_smoothen)
```

**Create a pencil sketch**

```python
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
final_img = dodgeV2(gray_image, img_smoothen)
cv2_imshow(final_img)
```
